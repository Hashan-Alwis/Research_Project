```python
import pandas as pd
import numpy as np

# Assuming `data` is your existing DataFrame
num_rows = len(data)
categories = ["Adventure & Active", "Cultural & Heritage", "Luxury & Relaxation", "Budget Travelers", "Business & Work-Related"]

# Calculate the number of rows per category
rows_per_category = num_rows // len(categories)

# Create a list of categories repeated equally
balanced_categories = categories * rows_per_category

# If there are leftover rows, add random categories to make up the remainder
remainder = num_rows % len(categories)
if remainder > 0:
    balanced_categories += list(np.random.choice(categories, remainder, replace=False))

# Shuffle the list to distribute randomly
np.random.shuffle(balanced_categories)

# Assign the balanced categories to the new column
data["Category"] = balanced_categories
```

```python
import pandas as pd
import numpy as np

# Define age ranges for each category
age_ranges = {
    "Adventure & Active": (18, 32),
    "Cultural & Heritage": (18, 70),
    "Luxury & Relaxation": (40, 70),
    "Budget Travelers": (18, 70),
    "Business & Work-Related": (25, 55)
}

# Function to generate age based on category, avoiding too many ages between 60-70
def generate_age(category):
    min_age, max_age = age_ranges[category]
    if max_age > 60:  # For ranges that include 60-70
        ages = list(range(min_age, max_age + 1))
        probabilities = [0.9 if age < 60 else 0.1 for age in ages]  # More weight for ages below 60
        probabilities = np.array(probabilities) / sum(probabilities)  # Normalize probabilities
        age = np.random.choice(ages, p=probabilities)
    else:
        age = np.random.randint(min_age, max_age + 1)
    return age

# Assign random ages based on category
data["Age"] = data["Category"].apply(generate_age)
```

```python
travel_style_probs = {
    "Adventure & Active": {"solo": 0.3, "couples": 0.1, "group": 0.6},
    "Cultural & Heritage": {"solo": 0.33, "couples": 0.33, "group": 0.34},
    "Luxury & Relaxation": {"solo": 0.4, "couples": 0.4, "group": 0.2},
    "Budget Travelers": {"solo": 0.7, "couples": 0.2, "group": 0.1},
    "Business & Work-Related": {"solo": 0.3, "couples": 0.0, "group": 0.7},
}

# Function to adjust probabilities based on age
def adjust_probs_by_age(category, age):
    base_probs = travel_style_probs[category]
    adjusted_probs = base_probs.copy()
    
    if age > 55:
        # Reduce probability of 'solo' and 'couples', increase 'group'
        if "solo" in adjusted_probs:
            adjusted_probs["solo"] *= 0.5
        if "couples" in adjusted_probs:
            adjusted_probs["couples"] *= 0.5
        if "group" in adjusted_probs:
            adjusted_probs["group"] += 0.5 * (adjusted_probs["solo"] + adjusted_probs["couples"])
    
    # Normalize probabilities to ensure they sum to 1
    total = sum(adjusted_probs.values())
    for key in adjusted_probs:
        adjusted_probs[key] /= total
    
    return adjusted_probs

# Function to assign travel style based on category and age
def assign_travel_style(row):
    category = row["Category"]
    age = row["Age"]
    adjusted_probs = adjust_probs_by_age(category, age)
    
    styles = list(adjusted_probs.keys())
    probabilities = list(adjusted_probs.values())
    return np.random.choice(styles, p=probabilities)

# Apply function to create "Travel Style" column
data["Travel Style"] = data.apply(assign_travel_style, axis=1)
```

```python
import pandas as pd
import numpy as np

# Define duration ranges (in days) based on category and travel style
duration_ranges = {
    "Adventure & Active": {
        "solo": (5, 10),
        "group": (10, 20),
        "couples": (7, 14)
    },
    "Cultural & Heritage": {
        "solo": (5, 14),
        "group": (7, 14),
        "couples": (7, 14)
    },
    "Luxury & Relaxation": {
        "solo": (5, 10),
        "group": (7, 14),
        "couples": (7, 14)
    },
    "Budget Travelers": {
        "solo": (5, 10),
        "group": (5, 8),
        "couples": (4, 9)
    },
    "Business & Work-Related": {
        "solo": (2, 5),
        "group": (2, 7)
    }
}

# Function to generate duration with extended ranges for certain categories
def assign_duration(row):
    category = row["Category"]
    travel_style = row["Travel Style"]
    
    if category in ["Cultural & Heritage", "Budget Travelers"]:
        # Use weighted probabilities for these categories
        probabilities = [0.6, 0.25, 0.1, 0.05]  # 60% standard, 25% 14-28, 10% 28-42, 5% 42-60
        ranges = [
            duration_ranges[category][travel_style],  # Standard range
            (14, 28),  # 14-28 days
            (28, 42),  # 28-42 days
            (42, 60)   # 42-60 days
        ]
        
        selected_range = ranges[np.random.choice(len(ranges), p=probabilities)]
        min_days, max_days = selected_range
        return np.random.randint(min_days, max_days + 1)
    
    else:
        # Use standard range for other categories
        min_days, max_days = duration_ranges[category][travel_style]
        return np.random.randint(min_days, max_days + 1)

# Apply function to create "Duration of Stay (Days)" column
data["Duration of Stay (Days)"] = data.apply(assign_duration, axis=1)
```

```python
def convert_to_month_week_day_format(days):
    # Calculate months, weeks, and days
    months = days // 30  # 1 month = 30 days
    days -= months * 30
    weeks = days // 7  # 1 week = 7 days
    days -= weeks * 7
    
    # Construct the duration string in "1m2w3d" format
    duration_str = f"{months}m{weeks}w{days}d"
    return duration_str

# Apply function to convert the "Duration of Stay (Days)" to "1m2w3d" format
data["Duration of Stay (Days)"] = data["Duration of Stay (Days)"].apply(convert_to_month_week_day_format)
```

```python
# Define income ranges based on category and age
income_ranges = {
    "Adventure & Active": {
        "18-30": (1500, 2500),
        "31-50": (2000, 3500),
        "51+": (3000, 4500)
    },
    "Cultural & Heritage": {
        "18-30": (2500, 3500),
        "31-50": (3000, 4500),
        "51+": (3500, 5500)
    },
    "Luxury & Relaxation": {
        "18-30": (4000, 6000),
        "31-50": (5000, 8000),
        "51+": (6000, 10000)
    },
    "Budget Travelers": {
        "18-30": (1200, 2000),
        "31-50": (1500, 2500),
        "51+": (2000, 3000)
    },
    "Business & Work-Related": {
        "18-30": (3500, 6000),
        "31-50": (4000, 7500),
        "51+": (5000, 8000)
    }
}

# Function to assign monthly income based on category and age
def assign_monthly_income(row):
    category = row["Category"]
    age = row["Age"]
    
    # Define age group based on age column
    if age <= 30:
        age_group = "18-30"
    elif 31 <= age <= 50:
        age_group = "31-50"
    else:
        age_group = "51+"
    
    # Get the income range for the specific category and age group
    min_income, max_income = income_ranges[category].get(age_group, (2000, 5000))
    
    # Randomly select an income within the range
    return np.random.randint(min_income, max_income + 1)

# Apply function to create "Monthly Income" column
data["Monthly Income"] = data.apply(assign_monthly_income, axis=1)
```

```python
import numpy as np
import pandas as pd



# Function to assign Tech Savviness based on Age and Category
def assign_tech_savvy(row):
    age = row["Age"]
    category = row["Category"]
    
    # Assign based on Age
    if age <= 35:
        tech_savvy = "High"
    elif age <= 50:
        tech_savvy = "Medium"
    else:
        tech_savvy = "Low"
    
    # Adjust based on Category
    if category == "Business & Work-Related":
        tech_savvy = "High"  # Business travelers are likely tech-savvy
    elif category == "Adventure & Active":
        if tech_savvy != "Low":
            tech_savvy = "Medium"  # Generally tech-savvy, but some may be less so
    elif category == "Luxury & Relaxation":
        if tech_savvy == "Low" and age < 50:
            tech_savvy = "Medium"  # Some luxury travelers may still be tech-savvy
    elif category == "Cultural & Heritage":
        if tech_savvy == "High":
            tech_savvy = "Medium"  # Generally, tech-savviness is medium for this category
    
    return tech_savvy

# Apply the function to assign the Tech Savvy column
data["Tech Savvy"] = data.apply(assign_tech_savvy, axis=1)

# Display the updated DataFrame

```

```python
def fill_nulls_with_random(df):
    """
    Fill NaN values in a DataFrame with random values from the same column.

    Parameters:
    df (pd.DataFrame): Input DataFrame with potential NaN values.

    Returns:
    pd.DataFrame: DataFrame with NaN values filled.
    """
    filled_df = df.copy()  # Create a copy to avoid modifying the original DataFrame

    for column in filled_df.columns:
        if filled_df[column].isnull().any():  # Check if the column has NaN values
            # Get the non-null values for the column
            non_null_values = filled_df[column].dropna()
            
            # Replace NaN values with random choice from the non-null values
            filled_df[column] = filled_df[column].apply(
                lambda x: np.random.choice(non_null_values) if pd.isnull(x) else x
            )

    return filled_df
```
