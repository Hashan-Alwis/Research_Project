```python
# Define the new values for the number_of_participants
new_participants = [2, 5, 8, 10]

# Create new rows based on existing single-person data
new_rows = []
for _, row in df.iterrows():
    for participants in new_participants:
        new_row = row.copy()
        new_row["number_of_participants"] = participants
        new_row["price(USD)"] = row["price(USD)"] * participants * 0.9  # Applying 10% discount
        new_rows.append(new_row)

# Append the new rows to the original DataFrame
new_rows_df = pd.DataFrame(new_rows)
df = pd.concat([df, new_rows_df], ignore_index=True)
```

```python
new_rows = []
for _, row in df.iterrows():
    # Make a deep copy to avoid reference issues
    week_day = row.copy()
    week_day["day_of_the_week"] = "week day"
    new_rows.append(week_day)
    weekend =row.copy()
    weekend["day_of_the_week"] = "weekend day"
    weekend["price(USD)"] = weekend["price(USD)"] * 1.2
    new_rows.append(weekend)

df_new_2 = pd.DataFrame(new_rows)
```

```python
new_rows = []
for _, row in df.iterrows():

    row_no_trainer = row.copy()
    row_no_trainer["with_trainer"] = "No"
    new_rows.append(row_no_trainer)
    

    if row.get("duration(h)", 0) == 5:
        row_with_trainer = row.copy()
        row_with_trainer["with_trainer"] = "Yes"
        row_with_trainer["price(USD)"] = row_with_trainer["price(USD)"] + 25  
        new_rows.append(row_with_trainer)

    if row.get("duration(h)", 0) == 7:
        row_with_trainer = row.copy()
        row_with_trainer["with_trainer"] = "Yes"
        row_with_trainer["price(USD)"] = row_with_trainer["price(USD)"] + 50  
        new_rows.append(row_with_trainer)

    if row.get("duration(h)", 0) == 3:
        row_with_trainer = row.copy()
        row_with_trainer["with_trainer"] = "Yes"
        row_with_trainer["price(USD)"] = row_with_trainer["price(USD)"] + 15  
        new_rows.append(row_with_trainer)

    if row.get("duration(h)", 0) == 72:
        row_with_trainer = row.copy()
        row_with_trainer["with_trainer"] = "Yes"
        row_with_trainer["price(USD)"] = row_with_trainer["price(USD)"] + 50  
        new_rows.append(row_with_trainer)

# Convert new rows to DataFrame
df_new = pd.DataFrame(new_rows)
```

```python
duration_mapping = {
    1: "one hour",
    2: "three hours",
    3: "six hours",
    5: "one day",
    7: "three day"
}

df['duration(h)'] = df['duration(h)'].map(duration_mapping)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numerical_columns = [ 'price(USD)']
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[col], palette="Set2")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Calculate IQR
Q1 = df['price(USD)'].quantile(0.25)
Q3 = df['price(USD)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['price(USD)'] >= lower_bound) & (df['price(USD)'] <= upper_bound)]
```
