```python
import pandas as pd
import random
import string

# Example DataFrame


# Function to generate a random unique ID of length 8
def generate_unique_id(existing_ids, length=8):
    while True:
        unique_id = ''.join(random.choices(string.ascii_uppercase + string.digits, k=length))
        if unique_id not in existing_ids:  # Ensure uniqueness
            return unique_id

# Create a set to track existing IDs
existing_ids = set()

# Add a new column with unique IDs
n['UserID'] = [generate_unique_id(existing_ids) for _ in range(len(n))]
existing_ids.update(n['UserID'])


```

```python
import pandas as pd
import numpy as np

# Define the categories and age ranges with probabilities
age_distribution = {
    'low level': [(18, 23, 1.0)],  # 100% in 18-23 range
    'low-mid level': [(20, 25, 0.8), (25, 28, 0.2)],
    'mid level': [(28, 32, 0.9), (32, 35, 0.1)],
    'mid-high level': [(32, 35, 0.4), (35, 45, 0.4), (45, 55, 0.2)],
    'high level': [(30, 40, 0.4), (40, 55, 0.6)],
}

# Function to generate a random age based on the category's distribution
def generate_age(category):
    ranges = age_distribution[category]
    # Separate the age ranges and probabilities
    age_ranges = [r[0:2] for r in ranges]
    probabilities = [r[2] for r in ranges]
    
    # Choose a range based on probabilities
    chosen_range = np.random.choice(len(age_ranges), p=probabilities)
    
    # Get the chosen age range
    min_age, max_age = age_ranges[chosen_range]
    
    # Generate a random age within the chosen range
    return np.random.randint(min_age, max_age + 1)

# Example DataFrame


# Add the age column
df['age'] = df['category'].apply(generate_age)

# print(df)
```

```python
import pandas as pd
import numpy as np

# Define the categories, positions, and age-based probabilities
category_positions = {
    'low level': ['Intern', 'Trainee', 'Junior Developer'],
    'low-mid level': ['Junior Developer', 'Junior Engineer', 'Associate Engineer'],
    'mid level': ['Associate Engineer', 'Software Engineer'],
    'mid-high level': ['Software Engineer', 'Senior Engineer', 'Tech Lead'],
    'high level': ['Senior Engineer', 'Tech Lead', 'Architect', 'Principal Engineer']
}

# Age-based probability distribution for each position
age_position_probability = {
    'low level': {
        (18, 23): {'Intern': 0.8, 'Trainee': 0.2, 'Junior Developer': 0.0},
    },
    'low-mid level': {
        (20, 25): {'Junior Developer': 0.5, 'Junior Engineer': 0.3, 'Associate Engineer': 0.2},
        (25, 28): {'Junior Developer': 0.3, 'Junior Engineer': 0.4, 'Associate Engineer': 0.3},
    },
    'mid level': {
        (26, 30): {'Associate Engineer': 0.4, 'Software Engineer': 0.6},
        (30, 32): {'Associate Engineer': 0.3, 'Software Engineer': 0.7},
    },
    'mid-high level': {
        (32, 35): {'Software Engineer': 0.3, 'Senior Engineer': 0.4, 'Tech Lead': 0.3},
        (35, 45): {'Senior Engineer': 0.5, 'Tech Lead': 0.3, 'Software Engineer': 0.2},
        (45, 55): {'Senior Engineer': 0.4, 'Tech Lead': 0.3, 'Software Engineer': 0.3},
    },
    'high level': {
        (30, 40): {'Senior Engineer': 0.3, 'Tech Lead': 0.4, 'Architect': 0.2, 'Principal Engineer': 0.1},
        (40, 55): {'Senior Engineer': 0.4, 'Tech Lead': 0.4, 'Architect': 0.1, 'Principal Engineer': 0.1},
    }
}

# Function to generate a random applying position based on category and age
def get_applying_position(category, age):
    # Choose position based on category
    positions = category_positions[category]
    
    if category in age_position_probability:
        # Find the age range that the person's age falls into
        age_ranges = age_position_probability[category]
        
        # Select the correct range based on age
        for (min_age, max_age), prob_dist in age_ranges.items():
            if min_age <= age <= max_age:
                # We have found the correct age range, now choose a position based on the probabilities
                if np.random.rand() < 0.5:  # 70% chance to consider age-based probabilities
                    # Select a position based on the age-specific probability distribution
                    chosen_position = np.random.choice(positions, p=[prob_dist.get(pos, 0) for pos in positions])
                    return chosen_position
                else:
                    # 30% chance to choose a random position based only on category
                    return np.random.choice(positions)

    # If no age-based probabilities, choose a random position based on category
    return np.random.choice(positions)

# Example DataFrame with category and age
# data = {'category': ['low level', 'low-mid level', 'mid level', 'mid-high level', 'high level'],
#         'age': [22, 26, 30, 40, 45]}  # Example ages
# df = pd.DataFrame(data)

# Add the "applying position" column based on both category and age
df['applying position'] = df.apply(lambda row: get_applying_position(row['category'], row['age']), axis=1)

# print(df)
```

```python
def fill_nulls_with_random(df):
    """
    Fill NaN values in a DataFrame with random values from the same column.

    Parameters:
    df (pd.DataFrame): Input DataFrame with potential NaN values.

    Returns:
    pd.DataFrame: DataFrame with NaN values filled.
    """
    filled_df = df.copy()  # Create a copy to avoid modifying the original DataFrame

    for column in filled_df.columns:
        if filled_df[column].isnull().any():  # Check if the column has NaN values
            # Get the non-null values for the column
            non_null_values = filled_df[column].dropna()
            
            # Replace NaN values with random choice from the non-null values
            filled_df[column] = filled_df[column].apply(
                lambda x: np.random.choice(non_null_values) if pd.isnull(x) else x
            )

    return filled_df
```
